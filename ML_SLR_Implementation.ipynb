```python
# to form a dataframe

data = pd.DataFrame(diabetes.data,columns =diabetes.feature_names )
```

```python
data
```

```python
data.head()
```

```python
# to make target
data['target'] = diabetes.target
```

```python
data
```

```python
# make independent and dependent var

x = data[['bmi']]
y = data['target']
```

```python
x
```

```python
y
```

```python
# Train test split

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# random state >> reproduce the same result again and again

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
x_train
```

```python
#  scaling
```

```python
# model training

from sklearn.linear_model import LinearRegression
```

```python
model =  LinearRegression()
```

```python
model
```

```python
x_train
```

```python
y_train
```

```python
# training of model

model.fit(x_train,y_train)
```

```python
# attribute of model

model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# visualize the result

plt.scatter(x_test,y_test,color = 'skyblue',label = 'Actual')

# to draw a best fit line
plt.plot(x_test,y_pred,color = 'red',label = 'Predicted')
plt.xlabel("BMI")
plt.ylabel("One year progression target ")
plt.title("Linear reg of diabetes data")
plt.legend()
plt.show()
```

```python
# model evaluation

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
